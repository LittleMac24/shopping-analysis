### Analyzing Categorical-Continous Relationships

- **Jitter** the 0’s and 1’s vertically so points don’t stack exactly on top of each other.
- **Fit** a logistic regression curve \(\Pr(Y=1\mid X=x\)=\frac{1}{1+e^{-(\beta_0+\beta_1x)}}\).
- **Reading it**:
    - A steep S-curve means \(X\) is a strong predictor.
    - The midpoint of the S-curve (\(\beta_0+\beta_1x=0\)) is the \(X\) value where \(\Pr(Y=1)=0.5\).

### Example of Strongly Right Skewed Data

- The distributions of these variables suffer immensely from their Right-Skewed (Geometric and Poisson) distributions.
- An Example below is given, being Administrative and Administrative Duration both having most of their sample have zero.

```python
plt.style.use('fivethirtyeight')

fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_figheight(5)
fig.set_figwidth(10)


sns.histplot(
    x = 'Administrative',
    data=df_shopper,
    stat='count',
    ax= ax1
)

sns.boxplot(x = 'Administrative',
            data=df_shopper,
            ax= ax2)

ax1.set_title('Admin Count Histogram', fontweight='bold')
ax2.set_title('Admin Count Boxplot', fontweight='bold')
```
